```python
import pandas as pd
data = pd.read_csv("D:/yahiya/workshop/car_purchasing.csv", encoding="ISO-8859-1")
```

```python
import seaborn as sns

sns.heatmap(data.corr(),annot=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.figure(figsize=(50, 50))

for i in range(len(data.columns) - 1):
    for j in range(i + 1, len(data.columns)):
        plt.subplot(len(data.columns) - 1, len(data.columns) - 1, i * (len(data.columns) - 1) + j)
        plt.scatter(data[data.columns[i]], data[data.columns[j]])
        plt.xlabel(data.columns[i],fontsize=40)
        plt.ylabel(data.columns[j],fontsize=40)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()

new_column = le.fit_transform(data["country"]) 
data.insert(loc=1, column="country_encoded", value=new_column)
```

```python
x=data.iloc[:,1:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression() 
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R-squared: {r2:.4f}")
```
